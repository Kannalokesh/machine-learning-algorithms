Here I considered Mean Squared Error(MSE)


```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
```

DATASET - 1

```python
df=pd.read_excel("LR_Share1.xlsx")
```

```python
x1=df.X
y1=df.Y
```

Splitting the data by using train_set_split method from sklearn.model_selection library

```python
x_train,x_test,y_train,y_test=train_test_split(x1,y1,train_size=0.8,test_size=0.2,random_state=100)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#finding the gradient descent and updating the learning parameters i.e m and b
def gradient(m1,b1,x1_train,y1_train,l):
    m_grad=0
    b_grad=0
    n=len(x1_train)
    for i in range(n):
        x=x1_train.iloc[i]
        y=y1_train.iloc[i]
        m_grad+=-(2/n)*x*(y-(m1*x+b1))
        b_grad+=-(2/n)*(y-(m1*x+b1))
    m=m1-m_grad*l
    b=b1-b_grad*l
    return m,b
```

```python
#defaukt values for m and b is 0
m=0
b=0
l=0.01       #learning rate
epochs=1000 

for i in range(epochs):
    m,b=gradient(m,b,x_train,y_train,l)
print(m,b)
```

plotting the graph for x_train and y_train

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,b+m*x_train,"red")
plt.show()
            
```

Plotting the graph for test data

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,b+m*x_test,"red")
plt.show()
```
